# Flagging Price Range

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply function to df
df_ex['price_range'] = df_ex.apply(price_label, axis=1)
```

```python
#better alternative to the function above using the main DF
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_orders_products_merged.loc[df_orders_products_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_orders_products_merged['price_range_loc'].value_counts(dropna = False)
```

# For-Loops Statement Practice

# Flagging Busiest Days

```python
#value of 0 represent Saturday
df_orders_products_merged['order_day_of_week'].value_counts(dropna=False)
```

```python
result = []

for value in df_orders_products_merged["order_day_of_week"]:
  if value == 0 or value==1:
    result.append("Busiest days")
  elif value == 3 or value==4:
    result.append("Least busy days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#combine loop statement with data set
df_orders_products_merged['busiest_days']=result
```

```python
df_orders_products_merged['busiest_days'].value_counts(dropna=False)
```

```python
#the amount of regular days and busy days are close in value
```

# Flagging Busiest Periods of the Day

```python
df_orders_products_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
#top 5 counts will be considered the 'busiest hours', bottom 5 count will be the 'least busiest hours', all else will be 'average hours'
```

```python
result2=[]

for value in df_orders_products_merged["order_hour_of_day"]:
    if value in [10,11,14,15,13]:
        result2.append("Most orders")
    elif value in [3,4,2,5,1]:
        result2.append("Fewest orders")
    else: 
        result2.append ("Average orders")
```

```python
result2
```

```python
df_orders_products_merged['busiest_period_of_the_day']=result2
```

```python
#flagging busiest times
df_orders_products_merged['busiest_period_of_the_day'].value_counts(dropna=False)
```
